```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Top 10 Revenue Giants in the Fortune 1000

```python
df1 = df[["Company","Revenues_M"]].sort_values(by = "Revenues_M", ascending=False).head(10)
df1
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x = "Company", y = "Revenues_M", data = df1, palette = "viridis")
plt.xticks(rotation = 45)
plt.show()
```

# Top 30 companies with the lowest Profits in the Fortune 1000

```python
#creating a new df to exclude NaN values
df3 = df.fillna(0)
x = (df3["Profits_M"] == 0)
df3 = df3.loc[~x]


df2 = df3.groupby("Company")["Profits_M"].mean().reset_index(name = "profit_m")


df2 = df2.sort_values(by = "profit_m", ascending = False).tail(30)


plt.figure(figsize = (8,6))

sns.barplot(x = "Company", y = "profit_m", data = df2, palette = "viridis")
plt.tight_layout()
plt.xticks(rotation = 90)

plt.show()
```

# Conclusion and Insights:

Conclusions and Insights

1- Revenue Giants in the Fortune 1000: The top three revenue-generating companies are Walmart, Amazon, and Apple. These companies significantly outpace others in terms of revenue, establishing themselves as leaders in their respective industries.

2- Companies with the Lowest Profits: Lumen Technologies leads the list of the 30 companies with the lowest profits, followed closely by 3M and Fidelity National Information Services. This highlights the challenges faced by these companies, despite their prominence.

3- Sector Representation by Company Count: Financials, Technology, and Energy sectors dominate the Fortune 1000, with the most companies represented. This indicates the broad reach and influence of these sectors in the corporate landscape.

4- Most Profitable Sectors: The Technology sector tops the list of the most profitable sectors, followed by Financials and Energy. This aligns with the dominance of these sectors in revenue generation and their pivotal role in the economy.

5- Companies with the Largest Workforces: Walmart, Amazon, and Home Depot are the top three employers among the Fortune 1000 companies. The significant workforce sizes reflect the operational scale and market reach of these retail giants.

6- Public vs. Private Companies: A vast majority (95.50%) of the Fortune 1000 companies are public, with only a small fraction being private. This underscores the importance of public markets for large-scale enterprises.

7- Preferred Headquarters Locations: New York City is the most preferred location for company headquarters, with 61 companies based there. Houston and Atlanta also emerge as popular choices, hosting 36 and 24 companies, respectively.

8- Profitability Among Fortune 1000 Companies: A large proportion of companies in the Fortune 1000 are profitable, indicating a generally healthy financial environment among these top firms.

9- Founder-Led Companies: Very few companies in the Fortune 1000 are still led by their founders. This suggests a trend where founders often step aside as companies grow and mature.

10- Female Leadership in the Fortune 1000: More than 80% of the companies are not led by female CEOs, highlighting the ongoing gender disparity in corporate leadership.

11- Profitability Comparison: Female-Led vs. Male-Led Companies: The analysis shows a comparison between female-led and male-led companies in terms of profitability. This comparison sheds light on the financial performance of companies based on the gender of their leadership.

12- Revenue vs. Profit Analysis with Profitability: The scatter plot reveals that companies with lower revenue are more likely to be loss-making, suggesting a correlation between revenue size and profitability.

13- Revenue vs. Profit Analysis with Founder-CEO Leadership: The scatter plot indicates the impact of having a founder as the CEO on revenue and profit, providing insights into the effectiveness of founder-led leadership in driving financial performance.

14- Revenue vs. Profit Analysis with Female CEO Leadership: This analysis explores how female leadership correlates with revenue and profit, offering a perspective on the financial success of companies with female CEOs.
